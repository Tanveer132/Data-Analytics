**Numpy Functions**

```python
import numpy as np
```

```python
#arange function in the numpy.
#arange([start,] stop[, step,], dtype=None)
array1=np.arange(1,10,2)
print(array1)
```

```python
#linspace function in the numpy
array2=np.linspace(1,10,50)
print(array2)
```

```python
array2d=array2.reshape((5,10))
print(array2d)
```

```python
 array3d=array2.reshape((2,5,5))
print(array3d)
```

```python
#ravel function in the numpy
#to convert multidimensional array into 1d array
ravel_array=array3d.ravel()
print(ravel_array)
```

```python
#flatten function in the numpy
#to convert multidimensional array into 1d array
flat_array=array3d.flatten()
print(flat_array)
```

```python
#transpose function in the numpy
#to convert dimensions of array
array2.transpose()
```
