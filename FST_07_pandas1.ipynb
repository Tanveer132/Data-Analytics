**pandas are used for Data Analytics**

```
#importing pandas 
import pandas as pd
```

```
dic={"roll":[1,2,3,4],"name":["a","b","c","d"],"marks":[46,64,78,87]}
df=pd.DataFrame(dic)
print(df)
```

```
print(df.roll)
print(df["roll"])
```

```
print(df.info)
```

```
#calculate mean
print(df.mean())
print(df.min())
print(df.max())
print(df.count())
print(df.describe())
```

```
#slicing using iloc
print(df.iloc[1,1])
print(df.iloc[:,:-1])
print(df.iloc[[1,3],:])
```

```
#slicing using loc
print(df.loc[1,"name"])
print(df.loc[[1,2,3],["name","marks"]])
```
