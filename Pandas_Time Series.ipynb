```python
#Creating the dataframe using dictionaries.
dic={
    "A": pd.Series(np.nan),
     "B":pd.Series([1,2,3],index=list(range(3)),dtype=float),
     "C":pd.date_range("20160812",periods=3),
     "D":np.array([3]*3),                          #length of array should be same as len of column
     "F":pd.Timestamp("20211212"),
     "G": pd.Categorical(["test", "train", "test"]),
}
df=pd.DataFrame(dic,index=dic["C"])
df
```

#Sorting

#datetime functionalities

```python
import datetime as dt
```

```python
#pass the argument in the form of (year,month,day,hr,min)
date=dt.datetime(2020,12,1)
print(date)
```

```python
date_list=["2018-12-12",np.datetime64("2018-12-12"),date]
print(type(date_list[0]))
print(type(date_list[1]))
print(type(date_list[2]))
```

```python
# convert strings, int, datetime in to datetime object using pandas.to_datetime
date_list=pd.to_datetime(date_list)
print(type(date_list[0]))
print(type(date_list[1]))
print(type(date_list[2]))
print(date_list)
```

#Resampling

```python
#create date range 
rng=pd.date_range("01/01/2012",periods=50,freq="30m")
print(type(rng))

#create random data and apply dates as index
data=pd.Series(np.random.randint(1,500,(len(rng))),index=rng)

#resample the data
data=data.resample("1h").sum()
data
```

#Timestamp

```python
# Performing date and time arithmetic with absolute or relative time increments
day=pd.Timestamp("19990307")
print(day)
print(day.day_name())
print(day.dayofweek)
print(day.days_in_month)
print(day.is_leap_year)
print(day.month_name())
```

#Assembling datetime from multiple DataFrame columns

```python
df=pd.DataFrame(
    {
        "year":[2000,2001,2002],
        "month":[1,2,3],
        "day":[23,13,12]
    }
)
print(df)
```

```python
dates=pd.to_datetime(df[["year","month","day"]], errors="coerce")  #coerce converts invalid data to NaT (not a time)
dates
```

#Origin parameter for the creation of datetime index

```python
#origin id the starting parameter
date=pd.to_datetime(np.arange(4),unit="D",origin="2021-07-03")
print(date)
```
